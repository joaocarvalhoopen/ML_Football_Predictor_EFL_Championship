```python
from pandas import ExcelWriter
from pandas import ExcelFile

loc_FLC_Standings = "./FLC_Standings/"

def get_last_year_team_name_position(dic_standings, team_name, year, empty_value = 22):
    key_str = str(year) + team_name.upper()
    if key_str in dic_standings:
        return dic_standings[key_str]
    else:
        return empty_value

def dic_union_2(dict1, dict2):
    return dict(list(dict1.items()) + list(dict2.items()))

def get_file_standings(filename, year, dic_standings, set_all_team_names):
    df = pd.read_excel(loc_FLC_Standings + filename, sheetname='Folha1') # sheetname='Sheet1')

    list_team_positions = df[df.columns[2]]
    #print(list_team_positions)
    list_team_positions = list_team_positions.tolist()
    #print(list_team_positions)
    
    for i, standings_team_name in enumerate(list_team_positions):
        print(i, standings_team_name)
        csv_team_name = convert_team_name_standing_to_CSV_name(standings_team_name)
        dic_standings[str(year) + csv_team_name.upper()] = i + 1
        set_all_team_names.add(csv_team_name)
        
    #print("Column headings:")
    #print(df.columns)    

    return dic_standings, set_all_team_names


def read_all_standings():
    standings = {}
    all_team_names = set()
    
    standings, all_team_names = get_file_standings('FLC_2004_2005.xlsx', 2004, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2005_2006.xlsx', 2005, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2006_2007.xlsx', 2006, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2007_2008.xlsx', 2007, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2008_2009.xlsx', 2008, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2009_2010.xlsx', 2009, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2010_2011.xlsx', 2010, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2011_2012.xlsx', 2011, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2012_2013.xlsx', 2012, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2013_2014.xlsx', 2013, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2014_2015.xlsx', 2014, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2015_2016.xlsx', 2015, standings, all_team_names)    
    standings, all_team_names = get_file_standings('FLC_2016_2017.xlsx', 2016, standings, all_team_names)    
            
    return standings, all_team_names

standings, all_team_names = read_all_standings()

print("all_team_names")
sorted_all_team_names = sorted(list(all_team_names)) #.sort()

print("sorted_all_teams_names: ")
print(sorted_all_team_names)

print()

print("last_year_team_position: ", get_last_year_team_name_position(standings, "Wolves", 2016))

print("last_year_team_position: ", get_last_year_team_name_position(standings, "Wolves", 2001))

# print("dic: ", standings)
```
