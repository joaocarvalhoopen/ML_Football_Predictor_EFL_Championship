```python
# Gets the goals scored agg arranged by teams and matchweek

##############################

##  JNC Version.
def get_gss_jnc(playing_stat):
    
    # print("Start: ")
    # display(playing_stat)
    
    playing_stat['HTGS'] = 0
    playing_stat['ATGS'] = 0
    playing_stat['HTGC'] = 0
    playing_stat['ATGC'] = 0
    
    team_names_set = set()
    goals_dic = {}
    
    for i in range(0, len(playing_stat)):
        HTgoals_scored = 0
        ATgoals_scored = 0
        HTgoals_conceded = 0
        ATgoals_conceded = 0
        
        end_loop_home_flag = False
        end_loop_away_flag = False
        
        if (i > 0):
            homeTeam = playing_stat.iloc[i].HomeTeam
            awayTeam = playing_stat.iloc[i].AwayTeam
            
            team_names_set.add(homeTeam)
            team_names_set.add(awayTeam)
            
            # Loop from the previous line to the line 0 (zero).
            for j in range(i-1, -1, -1):
            # for j in range(0, i-1):
                # Generate the goals_dic keys for the line for the home and away teams.
                dic_key_homeTeam = homeTeam + str(j)
                dic_key_awayTeam = awayTeam + str(j)
                
                # HomeTeam: If isn't in goals_dic (in cache) calculate the value, accumulating goals. 
                if (end_loop_home_flag == False):
                    if (dic_key_homeTeam in goals_dic):
                        end_loop_home_flag = True
                        HTgoals_scored_tmp, HTgoals_conceded_tmp = goals_dic[dic_key_homeTeam]
                        HTgoals_scored += HTgoals_scored_tmp
                        HTgoals_conceded += HTgoals_conceded_tmp
                    else:
                        if (homeTeam == playing_stat.iloc[j].HomeTeam):
                            HTgoals_scored   += playing_stat.iloc[j].FTHG
                            HTgoals_conceded += playing_stat.iloc[j].FTAG
                        elif (homeTeam == playing_stat.iloc[j].AwayTeam):
                            HTgoals_scored   += playing_stat.iloc[j].FTAG
                            HTgoals_conceded += playing_stat.iloc[j].FTHG

                # AwayTeam: If isn't in goals_dic (in cache) calculate the value, accumulating goals.     
                if (end_loop_away_flag == False):
                    if (dic_key_awayTeam in goals_dic):
                        end_loop_away_flag = True
                        ATgoals_scored_tmp, ATgoals_conceded_tmp = goals_dic[dic_key_awayTeam]
                        ATgoals_scored += ATgoals_scored_tmp
                        ATgoals_conceded += ATgoals_conceded_tmp
                    else:
                        if (awayTeam == playing_stat.iloc[j].HomeTeam):
                            ATgoals_scored   += playing_stat.iloc[j].FTHG
                            ATgoals_conceded += playing_stat.iloc[j].FTAG
                        elif (awayTeam == playing_stat.iloc[j].AwayTeam):
                            ATgoals_scored   += playing_stat.iloc[j].FTAG
                            ATgoals_conceded += playing_stat.iloc[j].FTHG
                
                if ((end_loop_home_flag == True) and (end_loop_away_flag == True)):
                    # End's the loop.
                    break
                            
            # For the Home and Away Teams, add the goals of the current position - 1 to the goals_dic.
            dic_key_homeTeam = homeTeam + str(i-1)
            dic_key_awayTeam = awayTeam + str(i-1)
            goals_dic[dic_key_homeTeam] = (HTgoals_scored, HTgoals_conceded)
            goals_dic[dic_key_awayTeam] = (ATgoals_scored, ATgoals_conceded)        

        #print("HTgoals_scored :", HTgoals_scored)            
         
        # ERRO: No atribui os valores s celulas do dataframe.    
        #playing_stat.iloc[i].HTGS = HTgoals_scored
        #playing_stat.iloc[i].ATGS = ATgoals_scored
        #playing_stat.iloc[i].HTGC = HTgoals_conceded
        #playing_stat.iloc[i].ATGC = ATgoals_conceded

        
        # Com esta soluo j atribui o valor s ceclulas correctamente.
        #raw_data_16.set_value(index=379,col='FTR',value='D')
        playing_stat.set_value(index=i, col='HTGS', value=HTgoals_scored)
        playing_stat.set_value(index=i, col='ATGS', value=ATgoals_scored)
        playing_stat.set_value(index=i, col='HTGC', value=HTgoals_conceded)
        playing_stat.set_value(index=i, col='ATGC', value=ATgoals_conceded)
        
        
     
    # Debug
    #print("After: ")
    #display(playing_stat)
    
    
    return playing_stat, team_names_set

##############################

# jnc comentei os dados para debug.


# Apply to each dataset
get_gss = get_gss_jnc

# TODO: Descomentar,  s para debug para ser mais rapido.
playing_statistics_1, team_names_set_1 = get_gss(playing_statistics_1)

playing_statistics_2, team_names_set_2 = get_gss(playing_statistics_2)
playing_statistics_3, team_names_set_3 = get_gss(playing_statistics_3)
playing_statistics_4, team_names_set_4 = get_gss(playing_statistics_4)
playing_statistics_5, team_names_set_5 = get_gss(playing_statistics_5)
playing_statistics_6, team_names_set_6 = get_gss(playing_statistics_6)
playing_statistics_7, team_names_set_7 = get_gss(playing_statistics_7)
playing_statistics_8, team_names_set_8 = get_gss(playing_statistics_8)
playing_statistics_9, team_names_set_9 = get_gss(playing_statistics_9)
playing_statistics_10, team_names_set_10 = get_gss(playing_statistics_10)
playing_statistics_11, team_names_set_11 = get_gss(playing_statistics_11)
playing_statistics_12, team_names_set_12 = get_gss(playing_statistics_12)
playing_statistics_13, team_names_set_13 = get_gss(playing_statistics_13)
playing_statistics_14, team_names_set_14 = get_gss(playing_statistics_14)

```

```python
# This function is used by the new (jnc) and by the old code.
def get_points(result):
    if result == 'W':
        return 3
    elif result == 'D':
        return 1
    else:
        return 0

############################
# JNC Version.

def get_cuml_points_jnc(match_res):
    # creates a dic of the points of each match by line.
    match_res_points = {}
    
    for team_line, matches in  match_res.items():
        accumulated_points = 0
        
        # calculate the points.
        for game_res in matches:
            accumulated_points += get_points(game_res)
            
        # Add the points to the line.
        match_res_points[team_line] = accumulated_points
    
    return match_res_points


def get_matchres_jnc(playing_stat, team_names_set):
    # NOTE: This line always have the resutls of the previous line.
    
    # Create a dictionary with team names with lines as keys.
    match_res = {}
    match_loc = {}
    match_line = {}
   
    # Set current line matches to 0 (zero).
    current_line_matches_res = {}
    current_line_matches_loc = {}
    current_line_matches_line = {}
    for team in team_names_set:
        current_line_matches_res[team] = []
        current_line_matches_loc[team] = []
        current_line_matches_line[team] = []
    
    for i in range(0, len(playing_stat)):
        homeTeam = playing_stat.iloc[i].HomeTeam
        awayTeam = playing_stat.iloc[i].AwayTeam
        
        # Add a copy of the matches the current results to the current line.
        for team in team_names_set:
            match_res[team + str(i)] = current_line_matches_res[team].copy()
            match_loc[team + str(i)] = current_line_matches_loc[team].copy()
            match_line[team + str(i)] = current_line_matches_line[team].copy()

            
        if playing_stat.iloc[i].FTR == 'H':
            current_line_matches_res[homeTeam].append('W')
            current_line_matches_res[awayTeam].append('L')
            
            current_line_matches_loc[homeTeam].append('H')
            current_line_matches_loc[awayTeam].append('A')
            
            current_line_matches_line[homeTeam].append(i)
            current_line_matches_line[awayTeam].append(i)
        
        elif playing_stat.iloc[i].FTR == 'A':
            current_line_matches_res[awayTeam].append('W')
            current_line_matches_res[homeTeam].append('L')
        
            current_line_matches_loc[homeTeam].append('H')
            current_line_matches_loc[awayTeam].append('A')
            
            current_line_matches_line[homeTeam].append(i)
            current_line_matches_line[awayTeam].append(i)
        
        elif playing_stat.iloc[i].FTR == 'D':
            current_line_matches_res[awayTeam].append('D')
            current_line_matches_res[homeTeam].append('D')
        
            current_line_matches_loc[homeTeam].append('H')
            current_line_matches_loc[awayTeam].append('A')
        
            current_line_matches_line[homeTeam].append(i)
            current_line_matches_line[awayTeam].append(i)
            
        else:
            # Caso em que no est preenchido!
            # Uitl para as linhas de resultado que se pretende prever.
            pass

    return match_res, match_loc, match_line


## JNC Version
def get_agg_points_jnc(playing_stat, team_names_set):
    
    # print("Start: ")
    # display(playing_stat)
    
    match_res, match_loc, match_line = get_matchres_jnc(playing_stat, team_names_set)
    cum_pts   = get_cuml_points_jnc(match_res)
    
    # Creates the 2 columns with the values 0 (zero).
    playing_stat['HTP'] = 0
    playing_stat['ATP'] = 0
        
    for i in range(0, len(playing_stat)):
        homeTeam = playing_stat.iloc[i].HomeTeam
        awayTeam = playing_stat.iloc[i].AwayTeam
           
        # Generate the goals_dic keys for the line for the home and away teams.
        dic_key_homeTeam = homeTeam + str(i)
        dic_key_awayTeam = awayTeam + str(i)

        HT_points = cum_pts[dic_key_homeTeam]
        AT_points = cum_pts[dic_key_awayTeam]
    
        playing_stat.set_value(index=i, col='HTP', value=HT_points)
        playing_stat.set_value(index=i, col='ATP', value=AT_points)

    return playing_stat, match_res, match_loc, match_line 

####################



# Apply to each dataset

# Changes the function to my new function.
#get_agg_points = get_agg_points_jnc


# print("Start: ")
# display(playing_statistics_1)
# playing_statistics_1 = get_agg_points(playing_statistics_1)
# print("After: ")
# display(playing_statistics_1)


# TODO: J est testado e est a calcular correctamente!!!!
#       Colocar descomentado este na verso final!!!!
#       Ter em ateno que ele  chamado mais  frente.

print("Start: ")
display(playing_statistics_1)
playing_statistics_1, match_res_1, match_loc_1, match_line_1 = get_agg_points_jnc(playing_statistics_1, team_names_set_1)
print("After: ")
display(playing_statistics_1)


playing_statistics_2, match_res_2, match_loc_2, match_line_2 = get_agg_points_jnc(playing_statistics_2, team_names_set_2)
playing_statistics_3, match_res_3, match_loc_3, match_line_3 = get_agg_points_jnc(playing_statistics_3, team_names_set_3)
playing_statistics_4, match_res_4, match_loc_4, match_line_4 = get_agg_points_jnc(playing_statistics_4, team_names_set_4)
playing_statistics_5, match_res_5, match_loc_5, match_line_5 = get_agg_points_jnc(playing_statistics_5, team_names_set_5)
playing_statistics_6, match_res_6, match_loc_6, match_line_6 = get_agg_points_jnc(playing_statistics_6, team_names_set_6)
playing_statistics_7, match_res_7, match_loc_7, match_line_7 = get_agg_points_jnc(playing_statistics_7, team_names_set_7)
playing_statistics_8, match_res_8, match_loc_8, match_line_8 = get_agg_points_jnc(playing_statistics_8, team_names_set_8)
playing_statistics_9, match_res_9, match_loc_9, match_line_9 = get_agg_points_jnc(playing_statistics_9, team_names_set_9)
playing_statistics_10, match_res_10, match_loc_10, match_line_10 = get_agg_points_jnc(playing_statistics_10, team_names_set_10)
playing_statistics_11, match_res_11, match_loc_11, match_line_11 = get_agg_points_jnc(playing_statistics_11, team_names_set_11)
playing_statistics_12, match_res_12, match_loc_12, match_line_12 = get_agg_points_jnc(playing_statistics_12, team_names_set_12)
playing_statistics_13, match_res_13, match_loc_13, match_line_13 = get_agg_points_jnc(playing_statistics_13, team_names_set_13)
playing_statistics_14, match_res_14, match_loc_14, match_line_14 = get_agg_points_jnc(playing_statistics_14, team_names_set_14)

```

```python
def only_hw(my_str):
    if my_str == 'H':
        return 'H'
    else:
        return 'NH'
    
def all_types_of_results(my_str):
    if my_str == 'H':
        return '0'
    elif my_str == 'D':
        return '1'
    else:
        return '2'
    
#playing_stat['FTR'] = playing_stat.FTR.apply(only_hw)
playing_stat['FTR'] = playing_stat.FTR.apply(all_types_of_results)

# Testing set (2015-16 season)
# playing_stat_test = playing_stat[5700:]
```
